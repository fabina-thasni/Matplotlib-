## Year wise sales of fruits

```python
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908 ]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
years = [2010, 2011, 2012, 2013, 2014, 2015]
```

### Year wise sales of oranges

```python
plt.figure(figsize=(10,6))
plt.plot(years,oranges,marker='*',color='orange')
plt.title('Year wise sales of oranges')
plt.xlabel('years')
plt.ylabel('oranges')
plt.legend('oranges')
plt.show()
```

### Year wise sales of apples

```python
plt.figure(figsize=(10,6))
plt.plot(years,yield_apples,marker='o',color='red')
plt.title('Year wise sales of apples')
plt.xlabel('years')
plt.ylabel('yeild_apples')
plt.legend('yeild_apples')
plt.show()
```

### comparing 2 plots

```python
plt.figure(figsize=(10,6))

plt.plot(years,oranges,marker='*',color='orange')
plt.plot(years,yield_apples,marker='o',color='red')

plt.title('Year wise sales of fruits')
plt.xlabel('years')
plt.ylabel('fruits')
plt.legend(['oranges','yeild_apples'])
plt.xticks(rotation=90)

plt.show()
```

# Box Plot

* A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). 
* It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical how tightly your data is grouped, and if and how your data is skewed.
* The IQR (the range between Q1 and Q3) forms the box, showing the spread of the middle 50% of the data.
* A smaller IQR indicatesss variability in the central data, while a larger IQR suggests more spread in the middle values. le

```python
windspeed=[10, 15, 8, 20, 12, 0, 7, 14, 21, 18, 9, 6, 3, 11, 5, 13, 17, 19, 22, 16, 8, 10, 7, 5, 13]
```

```python
plt.boxplot(windspeed)
plt.show()
```

# Violin plot

* A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data.
* It is used to visualize the distribution of numerical data.
* Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
windspeed=[10, 15, 8, 20, 12, 0, 7, 14, 21, 18, 9, 6, 3, 11, 5, 13, 17, 19, 22, 16, 8, 10, 7, 5, 13]
```

```python
plt.violinplot(windspeed)
plt.show()
```

# Scatter plot

* A scatter plot is a type of data visualization used to display the relationship between two numerical variables.
* Each point on the plot represents an observation, with its position determined by the values of the two variables.
* Scatter plots are useful for identifying trends, correlations, and patterns in data.

```python
np.random.seed(42)
```

```python
data={
    'feature X':np.random.normal(loc=50,scale=10,size=100),
    'feature Y':np.random.normal(loc=60,scale=15,size=100)
}
```

```python
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['feature X'],df['feature Y'],alpha=0.8,edgecolor='black')
plt.xlabel('feature X')
plt.ylabel('feature Y')
plt.title('scatter plot of features X,Y')
plt.show()
```

# pie chart
* A Pie Chart is a circular statistical plot that can display only one series of data. 
* The area of the chart is the total percentage of the given data.
* Pie charts are commonly used in business presentations like sales, operations, survey results, resources, etc. 

```python
# sales of a week
Units_Sold= [20, 35, 15, 40, 30]
Product= ['Product A', 'Product B', 'Product D', 'Product E', 'Product F']
```

```python
plt.pie(Units_Sold,labels=Product,shadow=True,explode=(0,0,0,0.1,0),autopct="%1.2f%%",startangle=90);
```

# Object-Oriented Approach Using Subplots

* In the object-oriented approach, we create a figure object that acts as an empty canvas and then we add a set of axes, or subplots, to it
* Another advantage of using the object-oriented plotting approach is the ability to create multiple subplots on a single figure object.
* In some situations, we want to compare different views of data side by side. We can accomplish this in Matplotlib using subplots

```python
import numpy as np
x=np.linspace(0,10,20)
```

```python
x
```

```python
y=x**2
```

```python
y
```

```python
y1=x**3
```

```python
y1
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Squared')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')

axes[1].plot(x, y1)
axes[1].set_title('x by x cubed')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('x by x Squared')
axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')

axes[1].plot(x, y1)
axes[1].set_title('x by x cubed')
axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')

plt.tight_layout()
plt.show()
```
